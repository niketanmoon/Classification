## Deleting Rows in `Embarked` and replacing the mean value for `Age`

```python
# Replace Age with mean
X_test["Age"] = X_test["Age"].fillna(X_test["Age"].mean())

# Replace the Fare with mean value
X_test["Fare"] = X_test["Fare"].fillna(X_test["Fare"].mean())

# Check again for missing value
X_test.isna().sum().max()
```

## Encoding the Categorical data

## First Feature Engineering the `Ticket` Parameter
- The problem with encoding Ticket is - since each ticket is different and some contains letters the encoding will be too much


## Taking care of `Tickets` data
- You may observe that if the tickets contains letters then there will be a space in between
- Let's split the each row ticket


```python
train_dataset["Ticket"] = train_dataset["Ticket"].apply(lambda x: float(x.split()[-1]) if len(x.split()) > 1 else 0)
X_test["Ticket"] = X_test["Ticket"].apply(lambda x: float(x.split()[-1]) if len(x.split()) > 1 else 0)

train_dataset["Ticket"].astype(str).astype(float)
X_test["Ticket"].astype(str).astype(float)
```

## Now taking all the columns and applying OneHotEncoding to Categorical Columns and MinMaxScaler technique to Numerical Columns

```python
# Applying one hot encoding on categorical cols and MinMaxScaling on Numerical columns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler


# getting the categorical columns
categorical_cols = train_dataset.select_dtypes(["object"]).columns

# getting the numerical columns
numerical_cols = train_dataset.select_dtypes(exclude=["object"]).columns
numerical_cols = numerical_cols[numerical_cols != "Survived"]

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop="first"), categorical_cols),
                                     ('scaler', MinMaxScaler(), numerical_cols),
                                    ], remainder='passthrough')
```

```python
X = train_dataset.drop(["Survived"], axis=1)
X = ct.fit_transform(X)
# after applying transformation
X.shape
```

```python
# Get the new column names after one-hot encoding
encoded_columns = ct.named_transformers_['encoder'].get_feature_names_out(categorical_cols)

# Combine with original numerical columns (not affected by encoding)
final_columns = list(encoded_columns) + list(numerical_cols)
final_columns
```

```python
# If you want to see value for first row
X[0]
```

All the numerical columns are converted in range of 0 to 1 and categorical columns are converted to binary

```python
## Applying the transformation on X_test
X_test = ct.transform(X_test)
X_test.shape
```
